```python

y6 = labels['nerves issues']
X_tr6, X_te6, y_tr6, y_ts6 = train_test_split(x2, y6, test_size=0.25, random_state=42,stratify=y6)


model1 = keras.Sequential([
    keras.layers.Input(shape=(18,)),  # Input layer with 16 features
    keras.layers.Dense(32, activation='relu'),  # Hidden layer with 32 units and ReLU activation
    keras.layers.Dense(1)  # Output layer with 1 unit (for regression)
])
model1.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
model1.fit(X_tr6,y_tr6, epochs=10, batch_size=64, validation_split=0.2)
y_pred = model1.predict(X_te6)
mse = mean_squared_error(y_ts6, y_pred)
print(f"Mean Squared Error: {mse:.2f}")

loss,accuracy = model1.evaluate(X_te6,y_ts6)
print(f"Accuracy: {accuracy:.2f}")
print(f"loss: {loss:.2f}")
```
